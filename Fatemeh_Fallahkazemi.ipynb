```python
#Charger le fichier Temperature 
import pandas as pd
from pandas import Series
temperature=pd.read_csv('C:/Users/Admin/Downloads/data_chapitre2(3)\data/temperatures.txt',header=None).squeeze()
temperature
```

```python
import pandas as pd

hours = [0, 3, 6, 9, 12, 15, 18, 21]
hour_data = hours * 91  # Répéter les horriares  pour 91 
temperature_data = temperature[:728]  

df = pd.DataFrame({'temp': temperature_data, 'hour': hour_data})
df
```

```python
#Question 3- calculer le moyen de temperature et la median
mean_temperature = df['temp'].mean()
median_temperature = df['temp'].median()
print("Le moyen de température est:",mean_temperature)
print("la médian de la temperature est:", median_temperature)
```

```python
#Question 4
import numpy as np

df.loc[(df['hour'] == 3) | (df['hour'] == 6), 'temp'] = np.nan
df
```

```python
#Question 5
import numpy as np

df.loc[df['hour'].between(3, 6, inclusive=True), 'temp'] = np.nan
df
```
