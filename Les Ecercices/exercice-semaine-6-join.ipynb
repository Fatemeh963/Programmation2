```python
#Question-5
bins=[df['trip_distance'].min(), 2, 10,df['trip_distance'].max()]
labels=['Short','Medium','Long']
df['trip_distance_group']=pd.cut(df['trip_distance'],bins,labels)
df
```

```python
df.groupby('trip_distance_group')['passenger_count'].mean().sort_values()
```

```python
df.groupby('trip_distance_group')['passenger_count'].mean().sort_values(ascending=False)
```

```python
#Question-6
df=df[df['trip_distance']!=0]
df['cost_km'] = df['total_amount'] / df['trip_distance']
df.sort_values('cost_km').tail(5)
```

```python

#Question-7
df=df[df['passenger_count']>=2]
df['amount_par_person'] = df['total_amount'] / df['passenger_count']
df.sort_values('amount_par_person').tail(10)
```
